```python
import os,sys,json
import core.nsf_utils, core.openaire_utils

import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
```

```python
data = [
    {"id": 1017296, "PIs": ["H. V. Jagadish"]},
    {"id": 540866, "PIs": ["Antony L. Hosking"]},
    {"id": 811691, "PIs": ["Jan Vitek", "Antony L. Hosking"]},
    {"id": 934218, "PIs": ["Dhiraj K. Sardar", "Igal Szleifer", "Fidel Santamaria", "Xomalin G. Peralta"]},
    {"id": 926127, "PIs": ["Jason Cong", "Jens Palsberg", "Glenn Reinman", "Vivek Sarkar", "Alex A. T. Bui"]},
    {"id": 1027253, "PIs": ["G. Philip Robertson", "Douglas A. Landis", "Thomas M. Schmidt", "Katherine L. Gross", "Stephen K. Hamilton"]}
]
```

```python
def plot_coauthor_net(data_idx, title_printout):
    dup_titles, total_titles, G = core.nsf_utils.publication_analysis(data[data_idx]["id"], title_printout=title_printout)
    print("\n* {}/{} duplicated papers are removed.".format(dup_titles, dup_titles+total_titles))
    print("* {} PIs, {}".format(len(data[data_idx]["PIs"]), data[data_idx]["PIs"]))
    pos = nx.spring_layout(G)
    nx.draw(G, pos)
    try:
        nx.draw_networkx_nodes(G, pos, nodelist=data[data_idx]["PIs"], node_color='r')
    except Exception as e:
        print(e)
    plt.show()
```

```python
plot_coauthor_net(0)
```

```python
plot_coauthor_net(1)
```

```python
plot_coauthor_net(2)
```
