```python
def plot_coauthor_net(award_id, title_printout=True):
    dup_titles, total_titles, G, pis = core.nsf_utils.publication_analysis(award_id, title_printout=title_printout)
    print("\n* {}/{} duplicated papers are removed.".format(dup_titles, dup_titles+total_titles))
    print("* {} PIs, {}".format(len(pis), pis))
    pos = nx.spring_layout(G)
    new_PI_list = []
    for pi in pis:
        if pi not in G.nodes:
            print("Node '{}' has no position.".format(pi))
        else:
            new_PI_list.append(pi)
    nx.draw(G, pos)
    try:
        nx.draw_networkx_nodes(G, pos, nodelist=new_PI_list, node_color='r')
    except Exception as e:
        print(e)
    plt.show()
```
